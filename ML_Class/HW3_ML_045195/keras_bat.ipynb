```python
import numpy as np
import matplotlib.pyplot as plt
from util import *

# -------------------------------------
# -------      import keras     -------
# -------------------------------------
import tensorflow as tf
from keras import Input
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.callbacks import EarlyStopping, ModelCheckpoint, LearningRateScheduler

class Config(object):
    def __init__(self, num_epoch=1000, batch_size=100, epochs_to_draw=10,
                 num_train=1000, activation='relu', loss='categorical_crossentropy',
                 opt='adam', learning_rate=0.001, beta1=0.9, beta2=0.999,
                 epsilon=1e-07, momentum=0):
        self.num_epoch = num_epoch
        self.batch_size = batch_size
        self.epochs_to_draw = epochs_to_draw
        self.num_train = num_train
        self.activation = activation
        self.learning_rate = learning_rate
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.momentum = momentum
        self.loss = loss
        self.opt = opt

def batch_train(cfg, accurate_plot):
    # Load data from file
    # Make sure that bat.dat is in data/
    train_x, train_y, test_x, test_y = get_bat_data()
    train_x, _, test_x = normalize(train_x, train_x, test_x)

    test_y  = test_y.flatten()
    train_y = train_y.flatten()
    num_class = (np.unique(train_y)).shape[0]

    # Pad 1 as the third feature of train_x and test_x
    train_x = add_one(train_x)
    test_x = add_one(test_x)

    # cfg = Config(num_epoch=2000, learning_rate=0.025, momentum_rate=0.9, num_train=train_x.shape[0], epochs_to_draw=250)

    # One hot encoding label
    train_set_x = train_x[:cfg.num_train].copy()
    train_set_y = train_y[:cfg.num_train].copy()
    train_set_y = create_one_hot(train_set_y, num_class)
    all_loss = []

    # Create NN classifier
    num_hidden_nodes = 100
    num_hidden_nodes_2 = 100
    num_hidden_nodes_3 = 100

    model = Sequential()
    model.add( Input( shape=(3,) ) )
    model.add( Dense( num_hidden_nodes, activation=cfg.activation ) )
    model.add( Dense( num_hidden_nodes_2, activation=cfg.activation ) )
    model.add( Dense( num_hidden_nodes_3, activation=cfg.activation ) )
    model.add( Dense( num_class, activation='softmax' ) )
    model.summary()

    # Define loss function, optimizer
    if cfg.opt == 'adam':
        model.compile(loss=cfg.loss, metrics=['accuracy'],
                      optimizer=tf.keras.optimizers.Adam(learning_rate=cfg.learning_rate,
                                                        beta_1=cfg.beta1, beta_2=cfg.beta2, 
                                                        epsilon=cfg.epsilon) )
        # Define callback
        callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10)
        # Start training
        H = model.fit(train_set_x, train_set_y, batch_size=cfg.batch_size, 
                      epochs=cfg.num_epoch, verbose=0, callbacks=[callback])
    elif cfg.opt == 'SGD':
        model.compile(loss=cfg.loss, metrics=['accuracy'],
                      optimizer=tf.keras.optimizers.SGD(learning_rate=cfg.learning_rate,
                                                        momentum=cfg.momentum) )
        # Define callback
        callback_lr = LearningRateScheduler(scheduler)
        callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10)
        # Start training
        H = model.fit(train_set_x, train_set_y, batch_size=cfg.batch_size, 
                      epochs=cfg.num_epoch, verbose=0, callbacks=[callback, callback_lr])
    else:
        model.compile(loss=cfg.loss, optimizer=cfg.optimizer, metrics = ['accuracy'])
        # Define callback
        callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10)
        # Start training
        H = model.fit(train_set_x, train_set_y, batch_size=cfg.batch_size, 
                      epochs=cfg.num_epoch, verbose=0, callbacks=[callback])
   
    # Draw figure
    fig = plt.figure()
    numOfEpoch = len(H.history['loss'])
    plt.plot(np.arange(0, numOfEpoch), H.history['loss'], label='training loss')
    # plt.plot(np.arange(0, numOfEpoch), H.history['val_loss'], label='validation loss')
    plt.plot(np.arange(0, numOfEpoch), H.history['accuracy'], label='accuracy')
    # plt.plot(np.arange(0, numOfEpoch), H.history['val_accuracy'], label='validation accuracy')
    plt.title('Accuracy and Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss|Accuracy')
    plt.show()
    y_hat = model.predict(train_x[0::3])
    print(y_hat.shape)
    visualize_point(train_x[0::3], train_y[0::3], y_hat)
    # plt.legend()
    test_y_one_hot = create_one_hot(test_y, num_class)
    # Evaluate model with test data set
    print("Test:")
    score = model.evaluate(test_x, test_y_one_hot, verbose=0)
    print(score)
    accurate_plot.append(score[1])

def scheduler(epoch, lr):
    if epoch > 100 and epoch < 110:
      return lr * tf.math.exp(-0.1)
    elif epoch > 275 and epoch < 305:
      return lr * tf.math.exp(-0.03)
    elif epoch > 425 and epoch < 450:
      return lr * tf.math.exp(-0.03)
    else:
      return lr

def bat_classification(epoch=5, act='relu'):
    # Define hyper-parameters and train-related parameters
    # 1. adam + categorical_crossentropy + relu
    # 2. adam + mean_squared_logarithmic_error + relu
    # 3. adam + poisson + relu
    # 4. SGD + categorical_crossentropy + relu
    # 5. SGD + poisson + relu

    cfg = Config(num_epoch=epoch, batch_size=1000, num_train=-1)
    accurate_plot = []
    # 1.
    print("1st parameters ...")
    cfg.learning_rate=0.0005
    cfg.activation=act
    batch_train(cfg, accurate_plot)

    # 2.
    print("2nd parameters ...")
    cfg.loss='mean_squared_logarithmic_error'
    batch_train(cfg, accurate_plot)

    # 3.
    print("3rd parameters ...")
    cfg.loss='poisson'
    batch_train(cfg, accurate_plot)

    #4.
    print("4th parameters ...")
    cfg.opt='SGD'
    cfg.learning_rate=0.02
    cfg.momentum=0.75
    cfg.loss='categorical_crossentropy'
    batch_train(cfg, accurate_plot)

    #5.
    print("5th parameters ...")
    cfg.learning_rate=0.03
    cfg.loss='poisson'
    batch_train(cfg, accurate_plot)

    # Plot accurracy
    plt.title('Accuracy')
    plt.xlabel('Parameters')
    plt.ylabel('Accuracy')
    plt.bar(range(len(accurate_plot)), accurate_plot, color ='maroon', width = 0.4)
    plt.ylim(0,1)
    plt.show()

    i = 0
    print("Accuracy: ")
    for acc in accurate_plot:
      i += 1
      print("%d. %f" % (i,acc) )

if __name__ == '__main__':
    plt.ion()
    epoch = 500
    bat_classification(epoch)
    bat_classification(epoch, act='tanh')
    bat_classification(epoch, act='sigmoid')
```
